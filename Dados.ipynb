   ## Importando os dados 
   Primeiro de tudo, a de se ficar claro que iremos utilizar o pacote `pandas` para fazer operações com dados em Data Frames. E essa utilização começa desde a primeira linha de código no python. Ah... antes disso, lembre-se de ter instalado o pacote pandas no seu computador através do comando `pip intall pandas` (assumo aqui que não preciso explicar sobre o Anaconda).

````python
import pandas as pd 
````

Vamos agora então importar os dados sobre **transações internacionais** de 2021-2022 conseguidos através do site [Kaggle](https://www.kaggle.com/datasets/shubhamptrivedi/world-trade-export-data20212022?select=WtoMetadata_20221014064149.csv), com fonte inicial da **Organização Mundial do Comércio**. O arquivo foi armazenado em um repositório público do GitHub, a sua importação foi feita da seguinte forma:

````python
url="https://github.com/talissonmeira/WTH/blob/main/WtoData.csv?raw=true"

df_wt = pd.read_csv(url,encoding='latin-1',low_memory=False)

#O encoding usado é declarado pois o padrão é o utf-8, diferente do qual está o arquivo.
#Como o arquivo não é tão grande podemos trabalhar com low_memory=False.
````

A tabela em questão é a seguinte:

````python
df_wt.head(10)
````

Para dar prosseguimento a análise precisamos obter também os dados de produção total do país (**GDP**) e esse dado vai ser obtido do [**World Bank**](https://data.worldbank.org/indicator/NY.GDP.MKTP.CD) com o mesmo código usado para obter os dados de transações internacionais e do mesmo repositório online do GitHub. Em seguida uma apresentação da tabela:

````python
url2="https://raw.githubusercontent.com/talissonmeira/WTH/main/gdp.csv"
df_gdp = pd.read_csv(url2)

df_gdp.head(10)
````

Por fim, vamos importar os dados de IDH (Human Development Index) do [Human Development Reports](https://hdr.undp.org/data-center/documentation-and-downloads), sob responsabilidade da **Organização das Nações Unidas**. O código também é colocado no repositório do GitHub, mas diferente dos outros, é um arquivo em formato `xlsx`, o que acarreta no comando ```pd.read.excel``` para ler o documento. A seguir a tabela: 

````python
url3='https://github.com/talissonmeira/WTH/blob/main/HDR21-22_Statistical_Annex_HDI_Trends_Table.xlsx?raw=true'
df_hdi = pd.read_excel(url3)

df_hdi.head(10)
````

## Limpeza e filtragem dos dados
Vamos começar agora os processos de escolha dos dados que nos interessam e limpeza dos países que não possuem todos os dados para ser feita a análise. O banco de dados mais limitante é o de transações internacionais, com diversos países sem dados e limitando em muito o escopo da análise. <br>
Vamos por partes, começando pelos dados de world trading, vamos observar os indicadores que temos diponíveis:

````python
df_wt['Indicator'].value_counts()
````

Como nossa comparação tem como objetivo a abertura comercial, escolhemos o lado do produtor para fazer a análise (escolha já declarada ao utilizar GDP) e precisamos apenas de um indicador de exportação total, para tal temos o `Total merchandise exports - monthly` e `Total merchandise exports - quarterly`. Por arbitrariedade escolhemos os dados referentes ao período trimestral. Assim criamos um novo data frame apenas com o indicador escolhido, para tal usamos o comando ```isin['elemento a ser buscado']```. Vamos usar apenas os dados do ano de 2021 por conta da disponibilidade de dados, então fazemos o mesmo para a coluna "Years":

````python
df_wt2 = df_wt[df_wt['Indicator'].isin(['Total merchandise exports - quarterly'])]
df_wt2 = df_wt2[df_wt2['Year'].isin([2021])]

df_wt2.head()
````

Aqui nos deparamos com um problema para a análise, a base de dados possui informações não só sobre países específicos mas sobre associações (por exeplo: MERCOSUL), porém, todos os países possuem um ISO3A Code, o que não é verdade para as associações, sabendo disso, usamos o comando ```.dropna``` para eliminar as linhas que tem valor nulo. A unica associação de países que possui o código é a União Europeia, como é só um valor, retiramos o mesmo com o o comando ```~df_em_questão['coluna']=='valor indesejado')]```:

````python
df_wt2 = df_wt2.dropna(subset=['Reporting Economy ISO3A Code'])
df_wt2= df_wt2[~(df_wt2['Reporting Economy ISO3A Code']=='EEC')]

df_wt2['Reporting Economy ISO3A Code'].value_counts()
````

Agora apenas eliminando as colunas irrelevantes para a análise: 

````python
keep=['Indicator','Reporting Economy ISO3A Code','Reporting Economy','Period Code','Unit','Value']
df_wt3 = df_wt2[keep].reset_index(drop=True)

df_wt3.head(10)
````

O próximo passo é efetuar a soma dos valores trimestrais para obter um valor anual, e transformar esses valores em dolares unitários (note que estava em milhares). <br>
Para tal, utilizamos o comando ```groupby[].sum``` que se encarrega de agrupar valores específicacos segundo classificações de colunas específicas:

````python
#Agrupando tudo e transformando os valores trimestrais em anuais
df_wt4=df_wt3.groupby(['Reporting Economy ISO3A Code','Indicator'],as_index=False)['Value'].sum()

#Eliminando a coluna Indicator
keep2=['Reporting Economy ISO3A Code','Value']
df_wt4=df_wt4[keep4]

#Trocando o nome da coluna Value e Reporting Economy ISO3A Code
df_wt4=df_wt4.rename(columns={'Value':'Exports'})
df_wt4=df_wt4.rename(columns={'Reporting Economy ISO3A Code':'Code'})

#Transformando de milhão para unidade 
df_wt4['Exports'] = df_wt4['Exports'] * 1000000

df_wt4.head(10)
````

Na tabela de GDP precisamos apenas dos dados de 2021 (estão em valores de dólares americanos unitários), deixamos a coluna Country Code pois vai ser utilizada para fazer operações mais a frente.

````python
keep3=['Country Name','2021','Country Code']
df_gdp2=df_gdp[keep3]

#Troca de nomes para facilitar entendimento e operações futuras
df_gdp2.rename(columns={'2021':'GDP'},inplace=True)
df_gdp2.rename(columns={'Country Code':'Code'},inplace=True)

df_gdp2.head(10)
````

O aviso quer dizer que estamos alterando um arquivo modificado e não o original, irrelevante para a operação em questão. <br>

Vamos agora para o ajuste da tabela de IDH:

````python
#Deixando apenas as colunas necessárias
keep4= ['Table 2. Human Development Index trends, 1990-2021', 'Unnamed: 16']
df_hdi2 = df_hdi[keep4]

#Alterando os nomes para otimização
df_hdi2.rename(columns={'Unnamed: 16':'HDI'},inplace=True)
df_hdi2.rename(columns={'Table 2. Human Development Index trends, 1990-2021':'Country Name'},inplace=True)

df_hdi2.head(10)
````

Em operações futuras os espaços em branco serão irrelevantes se não forem na tabela de world trading, portanto não foi uma preocupação eliminar os mesmos.

## Junção dos data frames para criação da tabela final
Vamos por partes, começaremos associando a tabela de transações internacionais com a tabela de produção total do país. Para tal iremos usar o comando ```pd.merge()``` que da maneira ```how='left'```utiliza os valores do data frame colocado a esquerda (df_wt4) como base e coloca os valores que tem o mesmo indicador (no caso código da economia) do data frame a esquerda (df_gdp2): 

````python
df_wt_gdp=pd.merge(df_wt4,df_gdp2,how='left',on=['Code'])

df_wt_gdp.head(100)
````

Por fim, adicionamos o HDI da mesma forma, mas agora com base nos nomes inteiros dos países. Essa escolha foi estratégica pois os nomes na tabela de GDP eram mais parecidos em suas particularidades com a tabela de HDI do que com a de world trading. 

````python
df_final=pd.merge(df_wt_gdp,df_hdi2,how='left',on=['Country Name'])

df_final.head(100)
````

Eliminando valores nulos da tabela:

````python
df_final=df_final.dropna()

df_final
````

Incluimos uma coluna com a porcentagem de exportações sobre a produção total:

````python
df_final['exports_over_GDP'] = df_final['Exports'] / df_final["GDP"]

df_final
````

Por fim, exportamos o arquivo (está disponível no repositório do GitHub).

````python
df_final.to_csv('df_final', sep=',', encoding='latin-1', index=False)
````
