```python
url3='https://github.com/talissonmeira/WTH/blob/main/HDR21-22_Statistical_Annex_HDI_Trends_Table.xlsx?raw=true'
df_hdi = pd.read_excel(url3)
```

```python
df_hdi.head(150)
#dados de uma tabela do excel, por isso muitos espaços nulos 
# ano de 2021
#tirado do Human Delepment Reports da UNDP (united nations development programme)
```

```python
columns_to_keep = ['Table 2. Human Development Index trends, 1990-2021', 'Unnamed: 16']
df_hdi2 = df_hdi[columns_to_keep]

df_hdi2.head()
```

a de se notar que há vários problemas na tabela, mas como iremos usar apenas os dados de 2021, não há problemas nisso.
daqui dá (e tem que) pegar vários detalhes da tabela (cabe gráficos)

```python
df_wt['Indicator'].value_counts()
#como só precisamos dos dados de exp total para entender a proporção de produção:
#os dados que nos servem são Total merchandise exports  (usar quarterly)
```

```python
df_wt2 = df_wt[df_wt['Indicator'].isin(['Total merchandise exports - quarterly'])]
df_wt2.head()
```

```python
#agora a gente precisa filtrar os dados pra apenas 2021 e apenas de economias nacionais
#exluir 2022 e exluir 'world' 'europe' 'asia'
df_wt2 = df_wt2[df_wt2['Year'].isin([2021])]
df_wt2.head()
```

```python
#agora pra excluir economias que não são países específicos:
#usar o Reporting Economy ISO3A Code	 que é nulo quando não é de país
#retirar a união europeia que é a única com código específico

df_wt2 = df_wt2.dropna(subset=['Reporting Economy ISO3A Code'])
df_wt2= df_wt2[~(df_wt2['Reporting Economy ISO3A Code']=='EEC')]
df_wt2.head()
```

```python
df_wt2['Reporting Economy ISO3A Code'].value_counts()
```

```python
#agora é hora de eliminar as colunas irrelevantes para a análise
keep=['Indicator','Reporting Economy ISO3A Code','Reporting Economy','Period Code','Unit','Value']
df_wt3 = df_wt2[keep].reset_index(drop=True)
df_wt3
```

```python
#agrupando tudo e transformando os valores quaterly em anual
#usei o nome inteiro do país porque é assim que ta nas outras tabelas
df_wt4=df_wt3.groupby(['Reporting Economy','Indicator'],as_index=False)['Value'].sum()
#aqui tem que trocar o nome dos total ... quarterly pra só total exports
df_wt4.head(100)
```

```python
#merging com a tabela de IDH
#trocando o nome da coluna na tabela de IDH pra pbater o merge
df_hdi2=df_hdi2.rename(columns={'Table 2. Human Development Index trends, 1990-2021':'Reporting Economy'})
df_wt_hdi=pd.merge(df_wt4,df_hdi2,how='left',on=['Reporting Economy'])
df_wt_hdi=df_wt_hdi.rename(columns={'Unnamed: 16':'HDI'})
df_wt_hdi.head(100)
```

```python
keep2=['Country Name','2021']
df_gdp=df_gdp[keep2]
df_gdp.rename(columns={'Country Name':'Reporting Economy'},inplace=True)
```

```python
df_final=pd.merge(df_wt_hdi,df_gdp,how='left',on=['Reporting Economy'])
df_final.head(100)
```
