```python
df_wt['Indicator'].value_counts()
#como só precisamos dos dados de imp e exp total para entender a balança comercial:
#os dados que nos servem são Total merchandise exports e import (usar quarterly)
```

```python
df_wt2 = df_wt[df_wt['Indicator'].isin(['Total merchandise exports - quarterly','Total merchandise imports - quarterly'])]
df_wt2.head()
```

```python
#agora a gente precisa filtrar os dados pra apenas 2021 e apenas de economias nacionais
#exluir 2022 e exluir 'world' 'europe' 'asia'
df_wt2 = df_wt2[df_wt2['Year'].isin([2021])]
df_wt2.head()
```

```python
#agora pra excluir economias que não são países específicos:
#usar o Reporting Economy ISO3A Code	 que é nulo quando não é de país
#retirar a união europeia que é a única com código específico

df_wt2 = df_wt2.dropna(subset=['Reporting Economy ISO3A Code'])
df_wt2= df_wt2[~(df_wt2['Reporting Economy ISO3A Code']=='EEC')]
df_wt2.head()
```

```python
df_wt2['Reporting Economy ISO3A Code'].value_counts()
```
